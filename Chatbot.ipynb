```python
#importing os,asyncio, AsyncAnthropic, load-dotenv for use in code
import os
import asyncio
from google import genai
from dotenv import load_dotenv

#loading the .env file in system memory using load-dotenv funtion
load_dotenv()

#creating an AsyncAnthropic object using respective command and inputting api_key fetched using os.getenv() from system memory
client = genai.Client()

chat = client.aio.chats.create(model="model_name")

#generates response for user input using specified model and returns resulted response
async def get_response(prompt):
  res = await chat.send_message(prompt)
  return res.text

#taking input from the user
async def get_input():
  prompt=input("How can I help you ? ")
  return prompt

async def main():
  print("Welcome to Chatbot")
  print("To leave the chatbot, type 'exit' or 'quit' ")

  while True:
    user_prompt = await get_input()

    #checking to see if the user wants to stop the service
    if user_prompt.lower() in ["exit","quit"]:
      print("Thanks for using our service")
      print("Goodbye")
      break

    response = await get_response(user_prompt)

    print(response)
```

```python
if __name__ == "__main__":
  asyncio.run(main())
```
